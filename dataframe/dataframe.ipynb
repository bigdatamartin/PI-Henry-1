#### 1. Preparamos el environment importando librerías.

```python
import pandas as pd
import sqlalchemy as sql
```

#### 2. Importamos y creamos un DataFrame de cada dataset por separado.

```python
df_csv_amazon = pd.read_csv('https://raw.githubusercontent.com/bigdatamartin/PI01_DATA05/main/Datasets/amazon_prime_titles.csv')
df_csv_disney = pd.read_csv('https://raw.githubusercontent.com/bigdatamartin/PI01_DATA05/main/Datasets/disney_plus_titles.csv')
df_csv_hulu = pd.read_csv('https://raw.githubusercontent.com/bigdatamartin/PI01_DATA05/main/Datasets/hulu_titles.csv')
df_json_netflix = pd.read_json('https://raw.githubusercontent.com/bigdatamartin/PI01_DATA05/main/Datasets/netflix_titles.json')
```

#### 3. Agregamos una columna al final de cada DataFrame, conteniendo a que plataforma hace referencia.

```python
df_csv_amazon['platform'] = 'Amazon Prime Video'
df_csv_amazon.head()
```

```python
df_csv_disney['platform'] = 'Disney Plus'
df_csv_disney.head()
```

```python
df_csv_hulu['platform'] = 'Hulu'
df_csv_hulu.head()
```

```python
df_json_netflix['platform'] = 'Netflix'
df_json_netflix.head()
```

#### 4. Se concatena los DataFrames en uno solo

```python
df = pd.concat([df_csv_amazon, df_csv_disney, df_csv_hulu, df_json_netflix])
```

```python
df.rename(columns = {'description':'TEST'}, inplace = True)
```

```python
df.tail()
```

#### 5. Al observarse un error en los index, se realiza un reset index.

```python
df = df.reset_index(drop=True)
```

```python
df.tail()
```

#### 6. Cambiamos el 'show_id' para unificar criterio de identificación para todas las plataformas.

```python
df['show_id'] = df.index * 135
df.head()
```


#### 7. Exploramos y trabajamos los valores nulos.

```python
df.isnull().sum()
```

```python
df[['duration']] = df[['duration']].fillna(value='sin datos')
df['duration'].sample(10)
```

```python
df[['rating']] = df[['rating']].fillna(value='sin datos')
df['rating'].sample(10)
```

```python
df[['date_added']] = df[['date_added']].fillna(value='sin datos')
df['date_added'].sample(10)
```

```python
df[['country']] = df[['country']].fillna(value='sin datos')
df['country'].sample(10)
```

```python
df[['cast']] = df[['cast']].fillna(value='sin datos')
df['cast'].sample(10)
```

```python
df[['director']] = df[['director']].fillna(value='sin datos')
df['director'].sample(10)
```

```python
df.isnull().sum()
```

#### 8. Modelamos la data.
####    La columna 'cast' tiene mas de un valor separado por comas. Por lo tanto separamos una fila por actor, manteniendo los valores relacionados.

```python
df_modelado_actors = df.assign(cast=df.cast.str.split(",")).explode('cast')
df_modelado_actors = df_modelado_actors[['platform', 'release_year', 'cast']]
df_modelado_actors.head()
```

```python
import warnings
warnings.filterwarnings("ignore")

from sqlalchemy import create_engine
engine = create_engine('mysql+pymysql://root:Mies2019!!@localhost:3306/proyecto_individual1')
df_modelado_actors.to_sql('actors', engine, index=False)
```
